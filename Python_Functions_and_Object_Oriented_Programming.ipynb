Question 1.  What is the difference between a function and a method in Python?
             
             
             ->  1. Definition

Function:
A function is a block of code defined using the def keyword that performs a particular task. It can exist independently and is not associated with any specific object.

Method:
A method is a function that is associated with an object. It is defined inside a class and is used to perform operations on the data of that class.

2. Association with Objects

A function is independent and can be called directly by its name.

A method is always associated with an object and is called using the object name followed by a dot (.).

Example:

# Function
def greet(name):
    return "Hello " + name

print(greet("John"))

# Method
name = "john"
print(name.upper())

Here, greet() is a function, while upper() is a method of the string object name.

3. Definition Location

A function is defined outside any class.

A method is defined inside a class.

Example:

class Student:
    def display(self):
        print("Student details")

Here, display() is a method because it is defined inside the Student class.

4. The Use of self

Functions do not require self.

Methods inside a class must use self as the first parameter (in most cases) to refer to the current object.

Example:

class Car:
    def start(self):
        print("Car started")

self represents the instance of the class.

5. How They Are Called

Function: function_name(arguments)

Method: object_name.method_name(arguments)

Example:

len("Python")        # Function
"python".capitalize()  # Method
6. Purpose

Functions are used for general tasks that are not related to objects.

Methods are used to define the behavior of objects in Object-Oriented Programming (OOP).

7. Key Differences (Table Format)
Basis	Function	Method
Definition	Defined using def outside class	Defined inside a class
Association	Not associated with object	Associated with object
Calling	Called directly	Called using object
First Parameter	No self required	Uses self (for instance methods)
Example	print()	"abc".upper()
Conclusion

In summary, a function is an independent block of reusable code, whereas a method is a function that belongs to an object and operates on the data of that object. Methods are an important concept in Object-Oriented Programming, while functions are commonly used in procedural programming. Both help in writing modular, reusable, and organized Python programs.



 Question 2. Explain the concept of function arguments and parameters in Python.
             

             ->  1. Definition of Parameters and Arguments
Parameters

Parameters are the variables listed inside the parentheses in the function definition. They act as placeholders to receive values when the function is called.

Arguments

Arguments are the actual values passed to the function when it is called.

Example:
def greet(name):   # 'name' is a parameter
    print("Hello", name)

greet("PeeYush")   # "PeeYush" is an argument

Here:

name → Parameter

"PeeYush" → Argument

2. Importance of Parameters and Arguments

They make functions reusable.

They allow dynamic input.

They improve flexibility and modularity of programs.

They help avoid repetition of code.

3. Types of Function Arguments in Python

Python supports different types of arguments:

1️⃣ Positional Arguments

Arguments are passed in the same order as parameters.

def add(a, b):
    return a + b

print(add(5, 3))

Here:

5 is assigned to a

3 is assigned to b

Order is important.

2️⃣ Keyword Arguments

Arguments are passed using parameter names.

def student(name, age):
    print(name, age)

student(age=20, name="Rahul")

Order does not matter here.

3️⃣ Default Arguments

Parameters can have default values. If no argument is passed, the default value is used.

def greet(name="Guest"):
    print("Hello", name)

greet()
greet("Aman")

If no argument is given, "Guest" is used.

4️⃣ Variable-Length Arguments

Python allows passing a variable number of arguments.

(a) *args (Non-keyword variable arguments)
def total(*numbers):
    print(sum(numbers))

total(1, 2, 3, 4)

*numbers collects multiple arguments into a tuple.

(b) **kwargs (Keyword variable arguments)
def display(**data):
    print(data)

display(name="John", age=25)

**data collects arguments into a dictionary.

4. Difference Between Parameters and Arguments
Basis	Parameters	Arguments
Definition	Variables in function definition	Actual values passed to function
Position	Inside function declaration	Inside function call
Purpose	Receive data	Provide data
Example	def add(a, b)	add(5, 3)
5. Scope of Parameters

Parameters are local to the function. They cannot be accessed outside the function.

def test(x):
    print(x)

test(10)
# print(x) → Error (x is not defined outside)
6. Conclusion

In conclusion, parameters and arguments are essential components of functions in Python. Parameters act as placeholders in the function definition, while arguments are the actual values supplied during the function call. Python provides different types of arguments such as positional, keyword, default, and variable-length arguments, making functions flexible and powerful. Understanding these concepts is important for writing efficient, reusable, and modular Python programs.

Question 3.  What is the purpose of the return statement in a Python function?
             

             ->  In Python, the return statement is used inside a function to send a result back to the place where the function was called. It plays a very important role in making functions useful, reusable, and meaningful in programming.

1. Definition of return Statement

The return statement is used to:

End the execution of a function

Send a value (or values) back to the caller

Syntax:
def function_name(parameters):
    # statements
    return value
2. Why the return Statement is Important

The return statement:

Sends the result of computation back to the caller

Allows storing the result in a variable

Makes functions reusable

Stops execution of the function immediately

Helps in building modular programs

3. Example of return Statement
def add(a, b):
    return a + b

result = add(5, 3)
print(result)

Explanation:

a + b is calculated.

The result is returned.

The returned value is stored in result.

Output:

8
4. Difference Between return and print

Many students confuse return with print, but they are different.

Basis	return	print
Purpose	Sends value back to caller	Displays output on screen
Usage	Used inside function	Used anywhere
Reusability	Value can be stored and reused	Cannot reuse printed value directly

Example:

def square(x):
    return x * x

def square_print(x):
    print(x * x)

a = square(4)      # Can store value
b = square_print(4)  # Cannot store printed value
5. Returning Multiple Values

Python allows returning more than one value.

def calculate(a, b):
    return a + b, a - b

sum_value, diff_value = calculate(10, 5)
print(sum_value, diff_value)

Python automatically returns values as a tuple.

6. Returning Without a Value

If return is used without a value, it returns None.

def message():
    return

print(message())

Output:

None
7. How return Stops Function Execution

When Python encounters a return statement, it immediately stops executing the function.

def check():
    print("Start")
    return
    print("End")   # This will not execute

check()

Output:

Start
8. Conclusion

In conclusion, the return statement is a fundamental part of Python functions. It allows functions to send results back to the caller, makes code reusable, supports multiple return values, and stops function execution when needed. Without the return statement, functions would not be able to provide meaningful outputs for further use in a program. Therefore, understanding the purpose and proper use of the return statement is essential for writing efficient and modular Python programs.

Question 4.  Write a Python program to calculate the factorial of a number provided by
the user using a while loop


             -> 1. Introduction

The factorial of a number is the product of all positive integers less than or equal to that number. It is denoted by n!.

Mathematically:


n!=n×(n−1)×(n−2)×...×1

Example:

5! = 5 × 4 × 3 × 2 × 1 = 120

0! = 1 (by definition)

2. Algorithm

Step 1: Start
Step 2: Input a number from the user
Step 3: If number is negative, display error message
Step 4: Initialize factorial = 1
Step 5: Use a while loop to multiply numbers from 1 to n
Step 6: Display the factorial
Step 7: Stop

3. Python Program Using While Loop
# Program to calculate factorial using while loop

num = int(input("Enter a number: "))

# Check if the number is negative
if num < 0:
    print("Factorial does not exist for negative numbers.")
else:
    factorial = 1
    i = 1
    
    while i <= num:
        factorial = factorial * i
        i += 1
        
    print("The factorial of", num, "is", factorial)
4. Explanation of the Program

input() function is used to take user input.

The program checks if the number is negative.

If the number is positive or zero:

factorial is initialized to 1.

while loop runs from 1 to the given number.

Each number is multiplied with factorial.

Finally, the result is displayed.

5. Sample Output
Enter a number: 5
The factorial of 5 is 120
6. Time Complexity

The program runs the loop n times.
Therefore, the time complexity is O(n).

7. Conclusion

This program successfully calculates the factorial of a number using a while loop. It handles negative numbers properly and demonstrates the use of loops, conditional statements, and user input in Python. The factorial concept is widely used in mathematics, permutations, combinations, and probability calculations.
             



            

 Question 5. What is Object-Oriented Programming (OOP)?
             

             -> 1. Introduction

Object-Oriented Programming (OOP) is a programming paradigm that organizes software design around objects rather than functions and logic. An object is a real-world entity that contains both data (attributes) and behavior (methods).

OOP helps in writing modular, reusable, and well-structured programs. Python is an object-oriented programming language that supports OOP concepts.

2. Basic Concepts of OOP
1️⃣ Class

A class is a blueprint or template used to create objects. It defines the properties (variables) and behaviors (functions) that the objects will have.

Example:

class Student:
    def __init__(self, name):
        self.name = name

    def display(self):
        print("Student Name:", self.name)
2️⃣ Object

An object is an instance of a class. It represents a real-world entity.

Example:

s1 = Student("Rahul")
s1.display()

Here:

Student → Class

s1 → Object

3️⃣ Four Main Principles of OOP

OOP is based on four important principles:

1. Encapsulation

Encapsulation means wrapping data and methods into a single unit (class). It restricts direct access to some components.

Example:

class Bank:
    def __init__(self, balance):
        self.__balance = balance   # Private variable

It protects data from accidental modification.

2. Abstraction

Abstraction means hiding complex implementation details and showing only essential features.

Example:
When you drive a car, you only use steering and pedals without knowing the internal engine mechanism.

In Python, abstraction can be implemented using abstract classes.

3. Inheritance

Inheritance allows one class to inherit properties and methods from another class.

Example:

class Animal:
    def speak(self):
        print("Animal speaks")

class Dog(Animal):
    pass

d = Dog()
d.speak()

Here, Dog inherits from Animal.

4. Polymorphism

Polymorphism means "many forms." It allows the same function name to behave differently.

Example:

def add(a, b):
    return a + b

print(add(5, 3))       # Integer addition
print(add("Hi ", "All"))  # String concatenation

Same function works differently for different data types.

4. Advantages of OOP

Code reusability

Better data security

Easy maintenance

Modular programming

Real-world modeling

Reduces code duplication

5. Applications of OOP

Software development

Game development

Web applications

Mobile applications

GUI-based applications

6. Conclusion

In conclusion, Object-Oriented Programming (OOP) is a programming approach that uses classes and objects to design programs. It is based on four main principles: encapsulation, abstraction, inheritance, and polymorphism. OOP makes programs more organized, reusable, secure, and easier to maintain. Therefore, it is widely used in modern software development and is an essential concept in Python programming.

 Question 7. What is the difference between abstraction and encapsulation?
             

             -> 1. Introduction

In Object-Oriented Programming (OOP), abstraction and encapsulation are two important principles. Both help in data security and better program structure, but they serve different purposes. Understanding the difference between them is essential for writing efficient and secure programs.

2. What is Abstraction?

Abstraction means hiding complex implementation details and showing only the essential features of an object.

It focuses on what an object does, not how it does it.

Real-Life Example:

When you drive a car, you use the steering wheel, accelerator, and brake. You do not need to know how the engine works internally. The complex mechanism is hidden — this is abstraction.

Example in Python:
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def area(self):
        return 3.14 * 5 * 5

c = Circle()
print(c.area())

Here:

The method area() is declared but not implemented in the abstract class.

Implementation details are hidden from the user.

3. What is Encapsulation?

Encapsulation means wrapping data (variables) and methods (functions) into a single unit (class) and restricting direct access to some data.

It focuses on data hiding and protecting data from unauthorized access.

Real-Life Example:

A bank account hides its balance from direct access. You can only access it through methods like deposit or withdraw.

Example in Python:
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance  # Private variable

    def deposit(self, amount):
        self.__balance += amount

    def get_balance(self):
        return self.__balance

acc = BankAccount(1000)
acc.deposit(500)
print(acc.get_balance())

Here:

__balance is private.

Direct access is restricted.

Data is accessed through methods.

4. Key Differences Between Abstraction and Encapsulation
Basis	Abstraction	Encapsulation
Meaning	Hiding implementation details	Hiding internal data
Focus	What an object does	How data is protected
Purpose	Reduce complexity	Protect data
Implementation	Using abstract classes and interfaces	Using access modifiers (private, protected)
Example	Abstract method	Private variable
5. Relationship Between Abstraction and Encapsulation

Encapsulation helps achieve abstraction.

Both improve code security and maintainability.

Both are fundamental principles of OOP.

6. Conclusion

In conclusion, abstraction hides complex implementation details and shows only necessary features, while encapsulation hides data and restricts direct access to it. Abstraction focuses on functionality, whereas encapsulation focuses on data security. Both principles play an important role in developing secure, modular, and maintainable object-oriented programs.

 Question 9. What is inheritance in OOP? Provide a simple example in Python.
             

             -> 1. Introduction

Inheritance is one of the fundamental principles of Object-Oriented Programming (OOP). It allows one class (called the child or derived class) to acquire the properties and methods of another class (called the parent or base class).

In simple words, inheritance allows code reusability by enabling a new class to use the features of an existing class.

2. Definition of Inheritance

Inheritance is a mechanism in which one class inherits attributes (variables) and methods (functions) from another class.

Parent Class (Base Class) → The class whose properties are inherited.

Child Class (Derived Class) → The class that inherits properties.

3. Why Inheritance is Important

Promotes code reusability

Reduces code duplication

Makes code easier to maintain

Supports hierarchical classification

Improves program structure

4. Syntax of Inheritance in Python
class ParentClass:
    # Parent class code

class ChildClass(ParentClass):
    # Child class code

The child class is written inside parentheses with the parent class name.

5. Simple Example in Python
# Parent class
class Animal:
    def speak(self):
        print("Animal makes a sound")

# Child class inheriting from Animal
class Dog(Animal):
    def bark(self):
        print("Dog barks")

# Creating object of Dog class
d = Dog()

# Calling methods
d.speak()   # Inherited method
d.bark()    # Own method
6. Explanation of the Example

Animal is the parent class.

Dog is the child class that inherits from Animal.

The Dog class automatically gets access to the speak() method.

Dog also has its own method bark().

When we create object d, we can call both methods.

Output:
Animal makes a sound
Dog barks
7. Types of Inheritance in Python

Single Inheritance

Multiple Inheritance

Multilevel Inheritance

Hierarchical Inheritance

Hybrid Inheritance

8. Conclusion

In conclusion, inheritance is an important feature of OOP that allows one class to reuse the properties and methods of another class. It promotes code reusability, reduces redundancy, and improves program organization. Through inheritance, Python programs become more efficient and easier to manage. The example of Animal and Dog clearly demonstrates how inheritance works in Python.

 Question 10. What is a lambda function in Python? Provide an example.
              

              -> 1. Introduction

A lambda function in Python is a small, anonymous (nameless) function that is defined using the lambda keyword. It is used to create simple functions in a single line without using the def keyword.

Lambda functions are mainly used when we need a short function for a short period of time.

2. Definition

A lambda function is an anonymous function that can have any number of arguments but only one expression.

Syntax:
lambda arguments: expression

lambda → Keyword used to define the function

arguments → Input parameters

expression → Single expression whose result is returned automatically

3. Features of Lambda Function

Anonymous (no function name)

Defined in a single line

Can take multiple arguments

Contains only one expression

Often used with functions like map(), filter(), and sorted()

4. Simple Example of Lambda Function
# Lambda function to add two numbers
add = lambda a, b: a + b

print(add(5, 3))
Output:
8

Explanation:

a and b are parameters.

a + b is the expression.

The result is automatically returned.

5. Lambda Function vs Normal Function
Normal Function:
def square(x):
    return x * x

print(square(4))
Lambda Function:
square = lambda x: x * x
print(square(4))

Both produce the same result, but the lambda function is shorter.

6. Example Using Lambda with map()
numbers = [1, 2, 3, 4]

squares = list(map(lambda x: x * x, numbers))

print(squares)
Output:
[1, 4, 9, 16]

Here, lambda is used to calculate squares of numbers in a list.

7. Advantages of Lambda Function

Reduces code length

Makes code cleaner and readable (for simple logic)

Useful for temporary functions

Saves time when writing short functions

8. Limitations

Can contain only one expression

Cannot include multiple statements

Not suitable for complex logic

9. Conclusion

In conclusion, a lambda function in Python is a small, anonymous function defined using the lambda keyword. It is useful for short and simple operations, especially when used with built-in functions like map() and filter(). Although it cannot handle complex logic, it helps in writing concise and efficient code. Therefore, lambda functions are an important feature in Python programming.
